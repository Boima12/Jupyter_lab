```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_map(ax):
    # =====================
    # Walls
    # =====================

    # Top wall
    ax.add_patch(
        patches.Rectangle(
            (-24.0/2, 8.5 - 0.2/2),
            24.0, 0.2,
            color='skyblue'
        )
    )

    # Bottom wall
    ax.add_patch(
        patches.Rectangle(
            (-24.0/2, -8.5 - 0.2/2),
            24.0, 0.2,
            color='skyblue'
        )
    )

    # Left wall
    ax.add_patch(
        patches.Rectangle(
            (-12.0 - 0.2/2, -18.0/2),
            0.2, 18.0,
            color='skyblue'
        )
    )

    # Right wall
    ax.add_patch(
        patches.Rectangle(
            (12.0 - 0.2/2, -18.0/2),
            0.2, 18.0,
            color='skyblue'
        )
    )

    # =====================
    # Obstacles
    # =====================

    ax.add_patch(patches.Circle((-5.0, 0.0), radius=0.2, color='red'))
    ax.add_patch(patches.Circle(( 5.0, 0.0), radius=0.2, color='red'))

    # =====================
    # Visual config
    # =====================

    ax.set_aspect('equal')
    ax.grid(True)
    ax.set_xlim(-13.5, 13.5)
    ax.set_ylim(-9.5, 9.5)
    ax.set_xlabel("X (meters)")
    ax.set_ylabel("Y (meters)")
```

```python
def load_waypoints(path="waypoints.txt"):
    pts = []
    with open(path) as f:
        for line in f:
            x, y = map(float, line.split())
            pts.append((x, y))
    return pts

# Load waypoints
wps = load_waypoints("waypoints.txt")
wx = [p[0] for p in wps]
wy = [p[1] for p in wps]

# Plot
plt.figure(figsize=(12, 9))
ax = plt.gca()

# Draw map
draw_map(ax)

# Draw waypoints
ax.plot(wx, wy, "go-", markersize=4, linewidth=1.5, label="Waypoints")

ax.legend()
ax.set_title("Map + Waypoints (Pure Pursuit)")
plt.show()
```
