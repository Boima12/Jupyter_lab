```python
class Node:
    def __init__(self, name):
        self.name = name
        self.g = 0
        self.h = 0
        self.f = 0
        self.parent = None

def a_star(graph, start, goal):
    open_set = [start] 
    closed_set = []

    while open_set:
        current_node = min(open_set, key=lambda node: node.f)

        if current_node.name == goal.name:
            path = []
            total_cost = current_node.g
            
            while current_node:
                path.append(current_node.name)
                current_node = current_node.parent
            return path[::-1], total_cost

        open_set.remove(current_node)
        closed_set.append(current_node)
        
        for neighbor_name, cost in graph[current_node.name]:
            neighbor = Node(neighbor_name)
            neighbor.g = current_node.g + cost
            neighbor.h = h(neighbor_name)
            neighbor.f = neighbor.g + neighbor.h
            neighbor.parent = current_node

            if neighbor in closed_set:
                continue
                
            existing_node_with_same_name = next((node for node in open_set if node.name == neighbor_name), None)
            
            if existing_node_with_same_name and neighbor.g >= existing_node_with_same_name.g:
                continue
                
            if existing_node_with_same_name:
                if neighbor.g < existing_node_with_same_name.g:
                    existing_node_with_same_name.g = neighbor.g
                    existing_node_with_same_name.f = neighbor.f
                    existing_node_with_same_name.parent = neighbor.parent
            else:
                open_set.append(neighbor)

    return None, 0

def tsp_with_astar(graph, start_name):
    unvisited = set(graph.keys()) - {start_name}
    current = start_name
    full_path = [current]
    total_cost = 0

    while unvisited:
        shortest = None
        shortest_path = []
        min_cost = float('inf')

        for city in unvisited:
            path, cost = a_star(graph, Node(current), Node(city))
            if path and cost < min_cost:
                shortest = city
                shortest_path = path
                min_cost = cost

        if not shortest:
            return None, float('inf')  # Failed to find path

        full_path += shortest_path[1:]  # Avoid re-adding current city
        total_cost += min_cost
        unvisited.remove(shortest)
        current = shortest

    # Return to start
    path_back, cost_back = a_star(graph, Node(current), Node(start_name))
    if not path_back:
        return None, float('inf')
    full_path += path_back[1:]
    total_cost += cost_back

    return full_path, total_cost
```
