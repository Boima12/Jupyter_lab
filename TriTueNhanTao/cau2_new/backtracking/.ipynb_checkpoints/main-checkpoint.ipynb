# Backtracking

### Setting up
no need to install any other libraries

### Bài toán Tô màu bản đồ

```python
# Hàm kiểm tra xem màu sắc có hợp lệ với các vùng kề không
def is_safe(graph, colors, node, color):
    for neighbor in graph[node]:
        if colors[neighbor] == color:
            return False
    return True

# Hàm quay lui để tô màu bản đồ
def solve_map_coloring(graph, m, colors, node=0):
    if node == len(graph):
        return True  # Tất cả các đỉnh đã được tô màu

    for color in range(1, m + 1):  # Thử từng màu
        if is_safe(graph, colors, node, color):
            colors[node] = color  # Gán màu cho đỉnh
            if solve_map_coloring(graph, m, colors, node + 1):
                return True
            colors[node] = 0  # Backtrack nếu không thành công

    return False  # Không tìm được cách tô hợp lệ

def map_coloring(graph, m):
    colors = [0] * len(graph)  # Ban đầu chưa tô màu nào
    if solve_map_coloring(graph, m, colors):
        return colors
    else:
        return None
```

### Sample 1

```python
graph = {
    0: [1, 2],
    1: [0, 2, 3],
    2: [0, 1, 3],
    3: [1, 2]
}

m = 3  # Số lượng màu cho phép
result = map_coloring(graph, m)

if result:
    print("Màu sắc của các vùng:", result)
else:
    print("Không thể tô màu bản đồ với", m, "màu.")
```
