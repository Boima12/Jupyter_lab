## Bài toán Tiên đoán giá nhà

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
```

```python
# Load data from file
data = np.loadtxt('priceHouse.txt', delimiter=',')

# Split into features (X) and target (y)
X = data[:, 0:2]  # First two columns: square footage and bedrooms
y = data[:, 2]    # Last column: price
```

```python
# Train the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Show coefficients
print("Trained Linear Regression Model:")
print(f"  Price = {model.coef_[0]:.2f} * sqft + {model.coef_[1]:.2f} * bedrooms + {model.intercept_:.2f}")
```

```python
# Predict price of a 2000 sqft, 3-bedroom house
prediction = model.predict([[2000, 3]])
print(f"\nPredicted price for 2000 sqft, 3-bedroom house: ${prediction[0]:,.2f}")
```

### Visualizing

```python
# 1️⃣ Actual vs Predicted Scatter Plot
y_pred = model.predict(X)
plt.figure(figsize=(6, 5))
plt.scatter(y, y_pred, color='blue', alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')  # Perfect prediction line
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Prices")
plt.grid(alpha=0.3)
plt.show()
```

```python
# 2️⃣ 3D Scatter + Regression Plane
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter points
ax.scatter(X[:, 0], X[:, 1], y, color='blue', alpha=0.6, label="Actual Data")

# Create grid for regression plane
sqft_range = np.linspace(X[:, 0].min(), X[:, 0].max(), 20)
bed_range = np.linspace(X[:, 1].min(), X[:, 1].max(), 20)
sqft_grid, bed_grid = np.meshgrid(sqft_range, bed_range)
price_grid = model.predict(np.c_[sqft_grid.ravel(), bed_grid.ravel()]).reshape(sqft_grid.shape)

# Plot surface
ax.plot_surface(sqft_grid, bed_grid, price_grid, color='orange', alpha=0.5)

ax.set_xlabel('Square Footage')
ax.set_ylabel('Bedrooms')
ax.set_zlabel('Price')
ax.set_title('3D Regression Plane')
plt.show()
```

```python
# 3️⃣ Feature vs Price (while fixing other feature)
fixed_bedrooms = 3
sqft_sorted = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
predicted_prices = model.predict(np.c_[
    sqft_sorted,
    np.full_like(sqft_sorted, fixed_bedrooms)
])

plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], y, color='blue', alpha=0.6, label="Actual Data")
plt.plot(sqft_sorted, predicted_prices, color='red', label=f"Bedrooms fixed at {fixed_bedrooms}")
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.title('Price vs Square Footage (Bedrooms Fixed)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
