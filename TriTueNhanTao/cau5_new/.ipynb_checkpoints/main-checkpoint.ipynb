## Bài toán Nhận dạng chữ số viết tay từ tập dữ liệu mnist.npz

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
```

```python
# Tải dữ liệu từ mnist.npz
with np.load('mnist.npz') as data:
    x_train = data['x_train']
    y_train = data['y_train']
    x_test = data['x_test']
    y_test = data['y_test']

# Chuẩn hóa dữ liệu về [0, 1]
x_train = x_train / 255.0
x_test = x_test / 255.0

# Encode nhãn
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
model = Sequential([
    Flatten(input_shape=(28, 28)),       # Chuyển ảnh 2D 28x28 thành 1D vector
    Dense(128, activation='relu'),       # Hidden layer 1
    Dense(64, activation='relu'),        # Hidden layer 2
    Dense(10, activation='softmax')      # Output layer
])
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.1)
```

```python
# Lấy ảnh đầu tiên test
idx = 0
sample = x_test[idx].reshape(1, 28, 28)

# Dự đoán
pred = model.predict(sample)
print("Dự đoán:", np.argmax(pred))

# Hiển thị ảnh
plt.imshow(x_test[idx], cmap='gray')
plt.show()
```
